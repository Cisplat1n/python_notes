#### Naming conventions in python: 

Rules for naming variables: 

* Can only start with a letter or _ (underscore) NOT a number
* Variables can only be alpha numeric (a-z, A-Z and 0-9 (the numbers can be as large as you want)) and _
* Variables are case sensitive (age, Age and AGE are all different)

Naming conventions: 

Some of the common naming schemes are camel case and snake case 

camel case: myVariableName

snake case: my_variable_name 

--- 

Always use descriptors when naming variables. For example if you are working with the age of an individual then maybe call the variable 'user_age'. Another good point is to avoid single letter names unless you have an actual practical use for them (x bad, x_ray_scan is fine). However, in a loop single letter names are perfectly acceptable.

```python
print('My favorite colors are', 'blue', 'green', 'red')
```

#### Common data types and how to get the 'Type' of a Variable:

A data type describes the kind of value a variable holds. For example, a number (int, float, etc.), a piece of text, or a list. 

Python is dynamically typed (Similar to JS), so it does NOT require you to declare the type for any variables. 

```python
name = 'John Doe' # Python knows this is a string
age = 25 # Python knows this is an integer
```

The real drawback of a dynamic interpreted language is that, unlike in a compiled language, the errors that can sometime be associated with type issues are only detected at run time when the code is actually be ran. A compiled language is more likely to detect an issue ahead of time, as it has to compile the code before it can even begin.

```python
# Some of the most common data types in Python include:
my_integer_var = 10
print('Integer:', my_integer_var) # Integer: 10

my_float_var = 4.50
print('Float:', my_float_var) # Float: 4.5 

my_string_var = 'hello'
print('String:', my_string_var) # String: hello

my_boolean_var = True
print('Boolean:', my_boolean_var) # Boolean: True

my_list_var = [1, 2, 3, 'four', 5.0]
print('List:', my_list_var) # List : [1, 2, 3, 'four', 5.0]

my_dictionary_var = {'name': 'Alice', 'age': 25}
print('Dictionary:', my_dictionary_var) # Dictionary: {'name': 'Alice', 'age': 25}

my_set_var = {7, 'hello', 8.5}
print('Set:', my_set_var) # Set: {7, 'hello', 8.5}

my_tuple_var = (7, 'hello', 8.5)
print('Tuple:', my_tuple_var) # Tuple: (7, 'hello', 8.5) 

my_range_var = range(5)
print('Range:', my_range_var) # Range: range(0, 5)

my_none_var = None
print('None:', my_none_var) # None: None
```

```python
# You can also use the type() function to check the type of a variable

number_var = 42
print(type(number_var)) # Output: <class 'int'>
```

A helpful function is the isinstance(). It allows you to compare a variable against a type and check a bool T/F check result based on if it passes or not. 

```python
print(isinstance('Hello', str)) # Output: True

print(isinstance(3.14, int)) # Output: False
```

#### Introduction to strings: 

Strings and string immutability, strings in python are declared using either "text" or 'text' (they are treated equally in python). To use a multi line string you can use """ text 
text
text""" (or use ')

```python
my_str_1 = 'Hello'
my_str_2 = "World" 

my_str_3 = """Multiline
string"""
my_str_4 = '''Another
multiline
string'''
```

If your string contains ' or ", simply use the opposite one to that to ensure that you do not cause issues (either cognitively or programmatically)

```python
msg = "It's a sunny day"
quote = 'She said, "Hello World!"'
```

Alternatively, you can use the \ as an escape character. 

```python
msg = 'It\'s a sunny day'
quote = "She said, \"Hello!\""
```

A useful technique to find out if a specific word is contained in a string is using the 'in' operator.

```python
my_str = "Hello World"

print('Hello' in my_str) # Output: True
```

It is also possible to get the length of a string.

```python
my_str = "Hello, World!"

print(len(my_str)) # Output: 13

my_str = "Hello World"

print(len(my_str)) # Output: 11
```

This leads on to working with indexing. Each position in a string can be directly accessed to print the associated value. The desired position is denoted by [] and starts from index position 0 (the first value in the index).

```python
my_str = "Hello World"

print(my_str[0]) # Output: H
print(my_str[6]) # Output: W
```

It is also possible to use negative values when accessing an index. This means it will start from the end of the string and move back the increment based on the negative value in the index.

```python
my_str = "Hello World"

print(my_str[-1]) # Output: d
print(my_str[-2]) # Output: l
```

Many programming languages group data types broadly as either primitve or reference types. Primitive types are simple and immutable, they cannot be changed once declared. Whereas reference types can hold multiple values, and can be either mutable or immutable. But Python doesn't draw a hard line between those two groups. Instead, all data gets treated as objects, and some objects are immutable while others are mutable.

Immutable data types can't be modified or altered once they are declared. You can point their variables at something new (reassignment) but you can't change the original object itself by adding, removing or replacing any of its elements. 

How does this apply to strings? 

Strings are considered immutable data types within python. 

```python
greeting = "Hello"
greeting = "Hi"
print(greeting) # hi
```

As we can see above, it did not concatenate the strings together. Initially, we assigned greeting to Hello and then assigned it to Hi. However, direct string modification is not supported (see below)

```python
greeting = "Hello"
greeting[0] = 'h' # This will raise an error because strings are immutable
```

Examples of other immutable data types in Python are integer, float, boolean, tuple, and range.

#### String Concatenation and Interpolation:

When working with strings, its common to want to concatenate them (add them together). We can do this using the + operator.

```python
my_str_1 = 'Hello'
my_str_2 = "World"

my_str_concat= my_str_1 + ' ' + my_str_2
print(my_str_concat) # Output: Hello World
```

This only works with strings in this context. Attempting to concatenate a string with an int will raise an error. 

```python
name = 'John Doe'
age = 26

my_str_concat= name + ' ' + age
print(my_str_concat) # Output: Hello World
```

This happens as python does not automatically convert between data types, i.e. it does not 'think' to change the 26 (int) to a 26 (str). What we can do though is use an operation to change the data type.

```python
name = 'John Doe'
age = 26

my_str_concat= name + ' ' + str(age)
print(my_str_concat) # Output: Hello World
```

It is also possible to use the augmented assignment operator for concatenation. This is represented by the +=, this performs both concatenation and assignment in a single step. 

```python
name = 'John Doe'
age = 26

name_and_age = name

name_and_age += ' ' + str(age)
print(name_and_age) # Output: John Doe 26
```

F strings:

These start with an f (f or F works) before the quotes, and allows you to embed variables or expressions inside replacement fields indicated by {}.

```python
name = 'John Doe'
age = 26

name_and_age = f"{name} is {age} years old."
print(name_and_age) # Output: John Doe is 26 years old.

#---------

num1 = 10
num2 = 5
print(f"the sum of {num1} and {num2} is {num1 + num2}") # Output: the sum of 10 and 5 is 15
```

Note how we didn't need to convert non-string types with the str(). With the example above, the value of the age, num1 and num2 variables in converted under the hood during the interpolation process.

#### String slicing

Like previous work above, we know a string can be accessed based on the index. We are also able to extract a specific portion of the string.

```python
# we can use string[start:stop] to get a substring
my_str = "Hello World"
print(my_str[1:4])
```

Note, the stop index is non-inclusive, this means it will take the string values from position 1, 2 and 3 BUT NOT 4 (see example above). It is also possible to omit the the start and stop indices of a string. If you omit the starting position, python will assume that you want to start from the beginning ([0]) or if you omit the end then python will assume you want to slice the string from whatever start position all the way to the end. 

You can also omit both the stop and start indices, this will cause python to extract the whole string.

```python
my_str = 'Hello world'
print(my_str[:7])  # Hello w


my_str = 'Hello world'
print(my_str[8:])  # rld

# string slicing does not affect the original string
my_str = 'Hello world'
substring = my_str[0:5]
print(substring)  # Hello
print(my_str)    # Hello world


my_str = 'Hello world'
print(my_str[:])  # Hello world
```

Apart from the start and stop indices, there's also an optional step parameter, which is used to specify the increment between each index in the slice.

'string[start:stop:step]'

```python
my_str = "Hello World"
print(my_str[0:11:2]) # Output: HloWrd
```

A minor tidbit is that you can reverse a string by leaving the start and stop empty but setting the step to a negative. 

```python
my_str= "Hello World"
print(my_str[::-1]) # Output: dlroW olleH
```
